```python
##Carregando dados limpos e prontos para análise

df = spark.table("workspace.default.vr_education_clean")
display(df)
```

```python
#Como está o engajamento dos alunos?
df.groupBy("engagement_level").count().orderBy("engagement_level").display()
```

```python
#Quanto mais VR, melhor o resultado?
df.groupBy("usage_of_vr_in_education") \
    .agg({"engagement_level": "avg"}) \
        .display()
```

```python
from pyspark.sql.functions import when, col, lower

df = df.withColumn(
    "improvement_numeric",
    when(lower(col("improvement_in_learning_outcomes")) == "yes", 1)
    .when(lower(col("improvement_in_learning_outcomes")) == "no", 0)
)
```

```python
#Existe relação entre VR e melhoria no aprendizado?
df.groupBy("improvement_in_learning_outcomes") \
    .count() \
        .display()

df.groupBy("usage_of_vr_in_education") \
    .agg({"improvement_numeric": "avg"}) \
        .display()
```

```python
##Instrutores mais preparados melhoram a eficacia do VR?

df.groupBy("instructor_vr_proficiency") \
    .agg({"perceived_effectiveness_of_vr": "avg"}) \
        .display()
```

```python
##VR aumenta criatividade e colaboração?

df.groupBy("impact_on_creativity").count().display()

df.groupBy("collaboration_with_peers_via_vr").count().display()
```

```python
##Regiões usam VR de forma diferente?

df.groupBy("region").count().display()

df.groupBy("region") \
    .agg({"engagement_level": "avg"}) \
        .display()
```
